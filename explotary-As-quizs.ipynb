```r
dat <- read.table('~/Downloads/household_power_consumption.txt', skip = 66637, nrows=2880, header = F, sep =';')
```

```r
colnames(dat) <- unlist(he)
```

```r
head(dat)
```

```r
hist(dat$Global_active_power,col="red",xlab="Global Active Power (kilowatts)",ylab="Frequency",main="Global Active Power")
```

```r
x <- paste(dat$Date, dat$Time)
#t2<-strptime(x, format="%m/%d/%Y %H:%M:%S")
t2<-as.POSIXct(x, format = '%d/%m/%Y %H:%M:%S')
class(t2[1])
```

```r
plot(t2,dat$Global_active_power,ylab="Global Active Power (kilowatts)",type="l")
```
